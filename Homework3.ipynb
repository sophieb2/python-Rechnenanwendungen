# Benutzte Formeln:
# $x_{n+1}=x_n-f(x_n)\frac{x_n-x_{n-1}}{f(x_n)-f(x_{n-1})}$
Die Nulstellen dieser Funktion werden gesucht:
# $(x+0.5)^n-1.5$
n=1, 2, 3, 4

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
def f(x, n):
    return ((x+0.5)**n)-1.5;
```

```python
#Nullstellensuchfunktion, die die reelle Nullstelle einer Funktion f(x) mittels des Sekantenverfahrens findet
def Nullstelle(f, x1, fmin):
    h_max=1
    it=0
    x2=x1+2*10**(-4)
    x3=x2
    while (np.abs(f(x3))>=fmin):
        it+=1
        #print(n, x0, x1, x2, f(x1), f(x2))
        x3=x2-f(x2)*(x2-x1)/(f(x2)-f(x1))
        if (x3-x2>h_max):
            x3=x2+1;
        if (x2-x3>h_max):
            x3=x2-1;
        x1=x2
        x2=x3
    return (x3, it)
```

```python
x=np.linspace(-2.5, 2.5, 1000000)
colors=['r','b','g','m']
plt.plot(x,0*x,color='k')
for n in range(4):
     plt.plot(x, f(x,n+1), color=colors[n], label="n="+str(n+1))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-10,10)
plt.legend(bbox_to_anchor = [1, 1])
```

```python
col_names = ["n", "fmin", "Nullstelle", "Anzahl der Funktionsaufrufe"]
data=[[]]

for n in range(4):
    a=Nullstelle(lambda x: f(x,n+1),0,10**(-7))
    data.append([n+1, 10**(-7), a[0], a[1]])
    a=Nullstelle(lambda x: f(x,n+1),0,10**(-11))
    data.append([n+1, 10**(-11), a[0], a[1]])
print(tabulate(data[1:], headers=col_names, tablefmt="pretty"))
```

```python
for n in range(4):
    for i in range(9):
        x0=i/2-2
        a=Nullstelle(lambda x: f(x,n+1),x0,10**(-11))
        plt.scatter(x0, a[1], color=colors[n])
plt.xlabel('Startwert')
plt.ylabel('Die Anzahl der Iterationen')
plt.title('n=1: rot; n=2: blau; n=3: grün; n=4: magenta; fmin=1e-11')
```
