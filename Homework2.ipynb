# Benutzte Formeln:
# $cos(cx)=\sum_{i=1}^{\infty}\frac{cos^{(i)}(0)(x-0)^i}{i!}$
# $cos^{(4k)}(cx)=c^i cos(cx) \Rightarrow cos^{(4k)}(0)=c^i$
# $cos^{(4k+1)}(cx)=-c^i sin(cx) \Rightarrow cos^{(4k+1)}(0)=0$
# $cos^{(4k+2)}(cx)=-c^i cos(cx) \Rightarrow cos^{(4k+2)}(0)=-c^i$
# $cos^{(4k+3)}(cx)=c^i sin(cx) \Rightarrow cos^{(4k+3)}(0)=0$

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
# Ableitungen von cos
def cosA(c,i):
    if i%4==0:
        return (c**i);
    elif i%4==1:
        return 0;
    elif i%4==2:
        return -(c**i);
    elif i%4==3:
        return 0;
```

```python
# Taylor-Funktion von cos
def cosT(x,c,n):
    ans=np.zeros(len(x))
    factorial=1
    for i in range(n+1):
        if(i>0):
            factorial=factorial*i;
        ans=ans + cosA(c,i) * (x**i) / factorial
    return ans;
```

```python
t0=time.perf_counter()
x=np.linspace(-10,10,1000)
c=0.5
T=0
plt.plot(x,np.cos(c*x), 'k--')
plt.xlabel('x')
plt.ylabel('cos(0.5x)')
plt.title('Approximation von cos(0.5x)')
plt.ylim(-2,2)
for n in range(6):
    t=time.perf_counter()
    plt.plot(x,cosT(x,c,2*n))
    print("n=",2*n, "c=", c, "Anzahl von Elementen in x:", len(x))
    t1=time.perf_counter()
    print("Die Laufzeit beträgt", t1-t, "Sekunden")
    T=T+t1-t
print("Mittlere Laufzeit:", T/6, "Sekunden")
print(time.perf_counter()-t0, "Sekunden hat das gesamte Programm gebraucht")
```
