Der Leapfrog Algorithmus:

$v_j(-\frac{\tau}{2})=v_j(0) - a_j(0)\frac{\tau}{2}$

$v_j(t+\frac{\tau}{2})=v_j(t-\frac{\tau}{2}) + a_j(t)\tau$

$r_j(t+\tau)=r_j(t) + v_j(t+\frac{\tau}{2}) \tau$

```python
def solveDGL (T, r, t_tot=10, tau=0.001): # in K, Angström, ps, ps
    #Initial values
    t=np.arange(0, t_tot, tau)
    r1=np.zeros((len(t)))
    r2=np.zeros((len(t)))
    v0_1=np.sqrt(T/(2*m))
    v0_2=-v0_1
    r1[0]=0
    r2[0]=r
    
    v_minus_half_tau_1=v0_1 - a(-r)*tau/2 #a(x)=Force(x)/m
    v_minus_half_tau_2=v0_2 - a(r)*tau/2
    for i in range(0, len(t)-1):
        #Atom 1:
        v_plus_half_tau_1=v_minus_half_tau_1 + a(r1[i]-r2[i])*tau #a(x)=Force(x)/m
        r1[i+1]=r1[i] + v_plus_half_tau_1*tau
        v_minus_half_tau_1=v_plus_half_tau_1
        #Atom 2:
        v_plus_half_tau_2=v_minus_half_tau_2 + a(r2[i]-r1[i])*tau
        r2[i+1]=r2[i] + v_plus_half_tau_2*tau
        v_minus_half_tau_2=v_plus_half_tau_2
        
        #print(i, r1[i+1], r2[i+1])
        
    return(t, r1, r2)
```

Trapezregel (n=1, 2 Stützpunkte):

$\int_a^b f(x) dx = \Sigma_{k=1}^N \frac{f(x_{k-1})+f(x_k)}{2}\Delta x_k =\frac{f(x_{0})+f(x_N)}{2} \Delta x + \Sigma_{k=1}^{N-1} f(x_k)\Delta x$

Die Simpson-Regel (n=2, 3 Stützpunkte):

$\int_a^b f(x) dx = \Sigma_{k=1}^N \frac{\Delta x_k}{3}(f(x_{k-1})+4f(x_{k-1}+h)+f(x_k))$

Boole-Regel (n=4, 5 Stützpunkte):

$\int_a^b f(x)dx = \Sigma_{k=1}^N \frac{2h}{45}(7f(x_{k-1})+32f(x_{k-1}+h)+12f(x_{k-1}+2h)+32f(x_{k-1}+3h)+7f(x_k))$

```python
#f - the function we're integrating, [a, b] - the range, n - N in the sums above
def integrate (f,a,b,n,method):
    aufrufe=0
    if(method=="trapez"):
        x=np.linspace(a,b,n+1)
        delta_x=x[1]-x[0]
        res=-f(x[0])-f(x[n])
        res/=2
        for i in range(n+1):
            res+=f(x[i])
            aufrufe+=1
        res=res*delta_x
    if(method=="simpson"):
        H=(b-a)/n #H=2h
        res=0
        for i in range(n):
            x=np.linspace(a+i*H,a+(i+1)*H,3)
            res+=f(x[0])+4*f(x[1])+f(x[2])
            aufrufe+=3
        res=res*H/6 # h=H/2
    if(method=="bode"):
        H=(b-a)/n #H=4h
        res=0
        for i in range(n):
            x=np.linspace(a+i*H,a+(i+1)*H,5)
            res+=7*f(x[0])+32*f(x[1])+12*f(x[2])+32*f(x[3])+7*f(x[4])
            aufrufe+=5
        res=res*H/90 # h=H/4
    return res, aufrufe
```
