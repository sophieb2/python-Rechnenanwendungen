```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sciopt
```

```python
def function(x, c=5, a=5):
    return a*np.sin(c*x);

def func2(x):
    return x**4-4*(x**3)-3*(x**2)+x
```

```python
choc=[]
for i in range(11):
    start_value=i/10
    zero=sciopt.fsolve(function, i, full_output="true")
    #print(i, zero[0], zero[1]["nfev"]) # nfev - number of iterations
    plt.scatter(zero[0], zero[1]["nfev"])
    #plt.scatter(zero[0], function(zero[0]))
    choc.append((zero[0], function(zero[0])))
```

```python
x=np.linspace(-np.pi, 4*np.pi, 100000)
plt.plot(x, function(x))
for i in range(len(choc)):
    plt.scatter(choc[i][0], choc[i][1]);
```
