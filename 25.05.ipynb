```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spin # from scipy import integrate as spin
```

```python
t = np.linspace(0,100,1000)
y0 = [0, 0]
m, k, b = 10, 10, 1
g=9.81
x_0=[2, 2]
```

```python
def function(z, t):
    return [z[1], -g-k*z[0]/m-b*z[1]/m];
```

```python
x=spin.odeint(function, x_0, t)
#x=x.transponse()
plt.plot(t,x[:,0])
```

```python
def derivatives(t, y, m, k, b, g=9.81):
    dydt = y[1]
    dvdt = -g-k*y[0]/m-b*y[1]/m
    
    return [dydt, dvdt]

t = np.linspace(0,100,1000)
y0 = [0, 0]
m, k, b = 10, 10, 1


# Look up documentation - as derivative is written, you will likely need an optional input called `tfirst`
res = spin.odeint(derivatives, y0, t, args=(m, k, b), tfirst="True")

plt.plot(t, res[:,0])
plt.plot(t, res[:,1])
```
