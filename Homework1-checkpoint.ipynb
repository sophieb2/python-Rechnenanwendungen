# Benutzte Formeln:
# $y=v_0sin(\theta_0)t-\frac{gt^2}{2}$
# $t=\frac{x}{v_0cos(\theta_0)}$

```python
import numpy as np
import matplotlib.pyplot as plt
import physics
from physics import _constantsDic as constants
from physics import Q
v_0 = 10+10*np.random.rand()
v_0 = Q(v_0, 'm/s')
h_0 = Q(0.49, 'm')
g = constants['g0']

x = np.arange(0, 50, 0.01)
x = Q(x, 'm')
theta_0=10*np.pi/180
t = x/(v_0*np.cos(theta_0))
y = h_0+v_0*np.sin(theta_0)*t-g*t*t/2
i=0
print("Anfangsgeschwindigkeit: ", v_0.value, v_0.unit)


maxwinkel=0
maxweite=-1
for i in range(10, 80, 1):
    theta_0=i*np.pi/180
    t = x/(v_0*np.cos(theta_0))
    y = h_0+v_0*np.sin(theta_0)*t-g*t*t/2
    nmax=0
    k=0
    while (y.value[k] >= 0) & (k < len(y.value)):
        nmax=k
        k=k+1;
    if x.value[nmax] > maxweite:
        maxwinkel=i
        maxweite=x.value[nmax];
    plt.plot(x.value[:nmax+1],y.value[:nmax+1])
    plt.xlabel('x, in m')
    plt.ylabel('y, in m')
    plt.title('Verschiedene Abwurfwinkel');
print("Abwurfwinkel, f√ºr den der Stein am Weitesten fliegt: ", maxwinkel, "Degrees")
print("Maximale Wurfweite: ", maxweite, x.unit)
```
